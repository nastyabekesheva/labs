```python
import matplotlib.pyplot as plt
import numpy as np

def draw_line(x1, x2, y1, y2, color):

    if x1 > x2: x = np.arange(x2,x1, step=0.05)[::-1]
    elif x1-x2==0 and y1-y2!=0: x = np.repeat(x1,abs(y1-y2)*20)
    else : x = np.arange(x1,x2, step=0.05)
    if y1 > y2: y = np.arange(y2,y1, step=0.05)[::-1]
    elif y1-y2==0 and x1-x2!=0: y = np.repeat(y1,abs(x1-x2)*20)
    else : y = np.arange(y1,y2, step=0.05)  
    if x1-x2==0 and y1-y2==0 : 
        x=x1
        y=y1

    plt.axis('off')
    plt.scatter(x,y,marker = '^', color = color)

draw_line(-0.2,-2,-0.2,-2,'black')
draw_line(0.2,2,0.2,2,'black')
draw_line(0,0,0,0,'blue')
draw_line(-2,-0.2,2,0.2,'red')
draw_line(0.2,2,-0.2,-2,'red')
draw_line(0,0,0.2,2,'green')
draw_line(0,0,-0.2,-2,'green')
draw_line(-0.2,-2,0,0,'yellow')
draw_line(0.2,2,0,0,'yellow')


#draw_line(0,2,0,2,'black')
```
