```python
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = plt.imread('/Users/nastyabekesheva/Desktop/labs/CM/image1.png')

# Create a copy of the image to apply the effect
modified_image = np.copy(image)

# Specify the region of interest (ROI) where you want to apply the effect
roi_x = 100  # x-coordinate of the top-left corner of the ROI
roi_y = 100  # y-coordinate of the top-left corner of the ROI
roi_width = 200  # width of the ROI
roi_height = 200  # height of the ROI

# Apply the effect to the ROI
effect_type = 'highlight'  # or 'darken', 'blur', etc.

if effect_type == 'highlight':
    # Increase the pixel values in the ROI to make it brighter
    modified_image[roi_y:roi_y + roi_height, roi_x:roi_x + roi_width, :] += 0.2
elif effect_type == 'darken':
    # Decrease the pixel values in the ROI to make it darker
    modified_image[roi_y:roi_y + roi_height, roi_x:roi_x + roi_width, :] -= 0.2
# Add other effects here

# Clip the modified pixel values to ensure they stay within the valid range of 0 to 1
modified_image = np.clip(modified_image, 0, 1)

# Display the original and modified images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(modified_image)
ax[1].set_title('Modified Image')
ax[1].axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = plt.imread('/Users/nastyabekesheva/Desktop/labs/CM/image1.png')

# Create a copy of the image to apply the effect
modified_image_bright = np.copy(image)
modified_image_dark = np.copy(image)

# Apply the brightening effect
modified_image_bright += 0.2

# Apply the darkening effect
modified_image_dark -= 0.2

# Ensure the pixel values are within the valid range of 0 to 1
modified_image_bright = np.clip(modified_image_bright, 0, 1)
modified_image_dark = np.clip(modified_image_dark, 0, 1)

# Display the original, brightened, and darkened images
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(modified_image_bright)
ax[1].set_title('Brightened Image')
ax[1].axis('off')
ax[2].imshow(modified_image_dark)
ax[2].set_title('Darkened Image')
ax[2].axis('off')
plt.show()
```
